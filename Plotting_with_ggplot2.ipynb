````r
# Fit separate Poisson regression models for each destination
destination_models <- by(Migration, Migration$destination, function(subdata) {
  glm(migrants ~ pops66, family = poisson, data = subdata)})


#3d: Create a plot with colored destinations and Poisson regression curves
ggplot(data = Migration, aes(x = pops66, y = migrants, color = destination)) +
  geom_point(alpha = 0.7) +
  geom_smooth(method = 'glm', family = poisson, se = FALSE, linetype = 'dashed', size = 1) +
  scale_color_viridis(discrete = TRUE) +
  labs(title = 'Relationship between Source Population and Migrants by Destination',
       x = 'Source Population in 1966',
       y = 'Number of Migrants') +
  theme_minimal() +
  theme(legend.position = 'bottom', legend.title = element_blank())
````

````r
# Initialize variables to store information
strongest_province <- NULL
strongest_effect <- 0

# Iterate over each destination model
for (i in seq_along(destination_models)) {
  # Extract coefficient for pops66
  coef_pops66 <- coef(destination_models[[i]])["pops66"]

  # Check if the absolute value of the coefficient is greater than the current strongest effect
  if (abs(coef_pops66) > strongest_effect) {
    strongest_effect <- abs(coef_pops66)
    strongest_province <- names(destination_models)[i]
  }
}

# Print the province with the strongest effect
cat("Province with the strongest effect:", strongest_province, "\n")
cat("Strongest effect of population on migrants:", strongest_effect, "\n")
````
